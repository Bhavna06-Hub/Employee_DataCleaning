```python
# Convert gender to consistent format
A['Gender'] = A['Gender'].replace({'M':'Male', 'F':'Female'})
A['Gender'] = A['Gender'].fillna('Unknown')
```

```python
# Convert all department to lowercase, then title case
A['Dept'] = A['Dept'].str.lower().str.title()
A['Dept'] = A['Dept'].fillna('Unknown')
```

```python
#standarize salary
# 1) Make everything a string
A['Salary'] = A['Salary'].astype(str)

# 2) Remove commas
A['Salary'] = A['Salary'].str.replace(',', '')

# 3) Handle "36k" type values
A['Salary'] = A['Salary'].str.replace('k', '000', case=False)

# 4) Convert to numeric
A['Salary'] = pd.to_numeric(A['Salary'], errors='coerce')

# 5) Fill missing with median
A['Salary'] =A['Salary'].fillna(A['Salary'].median())

A['Salary'] = A['Salary'].apply(lambda x: f"{int(x):,}")

```

```python
#standarize joining date
#  Convert to string and remove spaces
A['Joining_Date'] = A['Joining_Date'].astype(str).str.strip()

#  Replace empty or bad cells with 'Unknown'
A['Joining_Date'] = A['Joining_Date'].replace(['', 'nan', 'NaT', 'None'], 'Unknown')

#  Format readable dates; keep 'Unknown' as it is
A['Joining_Date'] = A['Joining_Date'].apply(
    lambda x: pd.to_datetime(x, errors='coerce').strftime('%d %b %Y') if x != 'Unknown' else 'Unknown'
)
```

```python
# Fill missing ages with median
A['Age'] = A['Age'].fillna(A['Age'].median())
```

```python
# Fill missing city with unknown
A['City'] =A['City'].fillna('Unknown')
```
